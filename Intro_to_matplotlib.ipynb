## Matplotlib example Workflow

```python
## 0 import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## 1 Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

## 2 Setup Plot
fig,ax = plt.subplots(figsize =(6,6)) ## (width,height)


## 3 Plotting Data
ax.plot(x,y)

## 4 Customize Plot
ax.set(title='Simple Plot', xlabel='x-axis',ylabel='y-axis')

## 5 Save and show you save the whole figure
fig.savefig('sample-plot.png')
```

```python
# There are two options for subplot
```

```python
## Option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to each different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
## Subplot Option #2
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot to each different axes
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which Method to use pyplot vs Matplotlib OO method
 * When Plotting something easy and fast use pyplot
 * When plotting something more advanced use OO method

```python
heart_disease.head()
```

```python
under_50 = heart_disease[heart_disease['age']<50]
under_50.head()
```

```python
under_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
over_50 = heart_disease[heart_disease['age']>50]
```

```python
## OO method
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
##ax.set_xlim([45,100])
```

```python
over_50.target.values
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
## Customise the plot
ax.set(title='Heart Disease and Cholestrol levels',xlabel="Age",ylabel="cholestrol");
ax.legend(*scatter.legend_elements(),title='Target');
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
## Subplot of chol,age and thalach
fig,(ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

## add data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
## customizing ax0
ax0.set(title="Heart Disease and Cholestrol levels",ylabel="Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(over_50["chol"].mean(),linestyle='--');


## add data to ax1
scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
##customizing ax1
ax1.set(title="Heart Disease and Thalach levels",xlabel="Age",ylabel="Thalach");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(over_50["thalach"].mean(),linestyle='--');

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
```
