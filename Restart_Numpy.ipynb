```python
np.random.seed(0)
sales_amounts = np.random.randint(20,size=(5,3))
sales_amounts
```

### Create a Dataframe

```python
weekly_sales = pd.DataFrame(sales_amounts,index=['MON','TUE','WED','THU','FRI'],columns=["Almond_butter","Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
price_array = np.array([8,10,12])
price_array
```

```python
price_array.shape
```

```python
butter_prices = pd.DataFrame(price_array.reshape(1,3),index=['Price'],columns=["Almond_butter","Peanut Butter","Cashew Butter"])
butter_prices
```

```python
total_sales = price_array.dot(sales_amounts.T)
total_sales
```

```python
butter_prices
```

```python
weekly_sales
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
weekly_sales["Total Earned '$'"] = daily_sales.T
```

```python
weekly_sales
```
